```java
import java.time.format.DateTimeFormatter;
import java.time.LocalDate;
import java.time.temporal.TemporalAdjusters;

class test{
    public static String getSystemInfo(String result, String actmode){

        String[] results = result.split("\n");
        String lastresult = results[results.length-1];
        String dtstring = lastresult.substring(0,22);

        System.out.println("actmode = " + actmode);
        LocalDate dt = LocalDate.parse(lastresult.substring(0,10), DateTimeFormatter.ofPattern("yyyy-MM-dd"));

        String retdatetime = "";
        DateTimeFormatter df = DateTimeFormatter.ofPattern("yyyy/MM/dd"); 

        switch(actmode){
            case "start_yesterday":
                retdatetime = dt.minusDays(1).format(df) + " 00:00:00";
                break;
            case "end_yesterday":
                retdatetime = dt.minusDays(1).format(df) + " 23:59:59";
                break;
            case "start_today":
                retdatetime = dt.format(df) + " 00:00:00";
                break;
            case "end_today":
                retdatetime = dt.format(df) + " 23:59:59";
                break;
            case "start_tommorow":
                retdatetime = dt.plusDays(1).format(df) + " 00:00:00";
                break;
            case "end_tommorow":
                retdatetime = dt.plusDays(1).format(df) + " 23:59:59";
                break;
            case "start_lastmonth":
                retdatetime = dt.minusMonths(1).with(TemporalAdjusters.firstDayOfMonth()).format(df) + " 00:00:00";
                break;
            case "end_lastmonth":
                retdatetime = dt.minusMonths(1).with(TemporalAdjusters.lastDayOfMonth()).format(df) + " 23:59:59";
                break;
            case "start_thismonth":
                retdatetime = dt.with(TemporalAdjusters.firstDayOfMonth()).format(df) + " 00:00:00";
                break;
            case "end_thismonth":
                retdatetime = dt.with(TemporalAdjusters.lastDayOfMonth()).format(df) + " 23:59:59";
                break;
            case "start_nextmonth":
                retdatetime = dt.plusMonths(1).with(TemporalAdjusters.firstDayOfMonth()).format(df) + " 00:00:00";
                break;
            case "end_nextmonth":
                retdatetime = dt.plusMonths(1).with(TemporalAdjusters.lastDayOfMonth()).format(df) + " 23:59:59";
                break;
        }

        System.out.println("return date = " + retdatetime);
        return "End";
    }
}

String result = "\n";
      result += "-------------------[xxxxxxxxxxxxxxxxxxxxxxxx]-----------------------\n";
      result += "2023-01-23 00:22:33.444 <ABCD - Text> 実行を開始しています...\n";
      result += "2023-01-24 00:33:44.555 <ABCD - Text> 実行を開始しています...\n";
      result += "2023-01-25 00:44:55.666 <ABCD - Text> 実行を開始しています...\n";
      result += "\n";

System.out.println("command result:\n" + result);

test.getSystemInfo(result,"start_yesterday");
test.getSystemInfo(result,"end_yesterday");
test.getSystemInfo(result,"start_today");
test.getSystemInfo(result,"end_today");
test.getSystemInfo(result,"start_tommorow");
test.getSystemInfo(result,"end_tommorow");
test.getSystemInfo(result,"start_lastmonth");
test.getSystemInfo(result,"end_lastmonth");
test.getSystemInfo(result,"start_thismonth");
test.getSystemInfo(result,"end_thismonth");
test.getSystemInfo(result,"start_nextmonth");
test.getSystemInfo(result,"end_nextmonth");
```
